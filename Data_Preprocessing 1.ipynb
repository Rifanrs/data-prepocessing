```python
df = df.drop(columns=['index', 'Revenue', 'Competitors'])
```

```python
df.head(5)
```

```python
# Mean imputation on Rating Column

avg_rating = df[df['Rating'] != -1]['Rating'].mean()

df['Rating'] = df['Rating'].apply(lambda x: avg_rating if x==-1 else x)
```

```python
# Mode imputation on Headquarters columns

mode_hq = df[df['Headquarters'] != '-1']['Headquarters'].mode()[0]

df['Headquarters'] = df['Headquarters'].apply(lambda x: mode_hq if x=="-1" else x)
```

```python
# Mode imputation on Size columns

mode_size = df[~df['Size'].isin(['-1','Unknown'])]['Size'].mode()[0]

df['Size'] = df['Size'].apply(lambda x: mode_size if x in ["-1", 'Unknown'] else x)
```

```python
# Mode imputation on Founded columns

mode_found = df[df['Founded'] != -1]['Founded'].mode()[0]

df['Founded'] = df['Founded'].apply(lambda x: mode_found if x==-1 else x)
```

```python
# Mode imputation on Type of ownership columns

mode_type = df[~df['Type of ownership'].isin(['-1','Unknown'])]['Type of ownership'].mode()[0]

df['Type of ownership'] = df['Type of ownership'].apply(lambda x: mode_type if x in ["-1", 'Unknown'] else x)
```

```python
# Mode imputation on Industry columns

mode_ind = df[df['Industry'] != '-1']['Industry'].mode()[0]

df['Industry'] = df['Industry'].apply(lambda x: mode_ind if x=="-1" else x)
```

```python
# Mode imputation on Sector columns

mode_sec = df[df['Sector'] != '-1']['Sector'].mode()[0]

df['Sector'] = df['Sector'].apply(lambda x: mode_sec if x=="-1" else x)
```

## Categorical Data Encoding

Categorical data encoding is the process of converting categorical variables into a format that can be used for machine learning models. Categorical variables are those that represent categories or groups, such as "red," "blue," "green" for a color variable or "male" and "female" for a gender variable. There are many ways to encode your categorical data, below are the encoding strategy that are often used in practice:

1. One-Hot Encoding
2. Label Encoding or Ordinal Encoding
3. Target Encoding

### One-Hot Encoding

```python
dummies = pd.get_dummies(df['Size'])

dummies = dummies.apply(lambda x: x.map({False: 0, True: 1}))
dummies
```

### Label Encoding

```python
def label_encoding_size_col(x):
  if x == '1 to 50 employees':
    return 0
  elif x == '51 to 200 employees':
    return 1
  elif x == '201 to 500 employees':
    return 2
  elif x == '501 to 1000 employees':
    return 3
  elif x == '1001 to 5000 employees':
    return 4
  elif x == '5001 to 10000 employees':
    return 5
  else:
    return 6
```

```python
df['Size'].apply(label_encoding_size_col)
```

### Job Title

```python
df['Job Title'].value_counts().head(50)
```

```python
def clean_job_title(x):
  if 'data scientist' in x.lower():
    return 'Data Scientist'
  elif 'data engineer' in x.lower():
    return 'Data Engineer'
  elif 'machine learning engineer' in x.lower():
    return 'ML Engineer'
  elif 'data analyst' in x.lower():
    return 'Data Analyst'
  elif 'machine learning scientist' in x.lower():
    return 'ML Scientist'
  elif 'business intelligence' in x.lower():
    return 'Business Intelligence'
  elif 'research scientist' in x.lower():
    return 'Research Scientist'
  elif 'data science manager' in x.lower():
    return 'Data Science Manager'
  else:
    return 'Others'
```

```python
df['Job Title'] = df['Job Title'].apply(clean_job_title)

df['Job Title'].value_counts()
```

### Salary Estimate

```python
df['Salary Estimate'].value_counts()
```

```python
def get_mean_salary(text):
  '''
    function to extract the mean of salary range
  '''
  numb = re.findall(r'[0-9]+', text) #get all numbers in pandas series
  numb = [int(x) for x in numb]
  return np.mean(numb) * 1000
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(get_mean_salary)

df['Salary Estimate'].value_counts()
```
